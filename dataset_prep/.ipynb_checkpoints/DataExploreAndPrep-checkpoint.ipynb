## Udacity Object Detections Dataset - Data Exploration

The udacity object detection dataset contains a labels.csv file that contains information of all ground truth bounding box and their corresponding labels for each image


NOTE: The original labels.csv from the tar file has an incorrect heading
    
    xmin	xmax	ymin	ymax	Frame	Label	Preview URL
    
The xmax and ymin labels are interchanged. Instead of handling this in code, we just updated the extracted csv dile header as follows.

    xmin	ymin	xmax	ymax	Frame	Label	Preview URL

I don't want to handle this in the code as the dataset will likely be corrected in the future.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import csv
import cv2

%matplotlib inline

## THESE should go into settings later
IMAGE_DIR = '../datasets/object-detection-crowdai/'
```

```python
# open the updated csv file and read the data

rows = []
with open('../datasets/object-detection-crowdai/labels.csv') as f :
    # read first 3 lines and use csv.Sniffer to check if there is a header
    has_header = csv.Sniffer().has_header(f.readline() + f.readline() + f.readline())
    # reset file ptr to start of file
    f.seek(0)
    
    reader = csv.reader(f)
    if has_header :
        print ('header: \n' + str(next(reader, None)))
        print("Skipping header")
    
    for row in reader :
        rows.append(row)

len(rows)
```

#### Define Bounding Box and Annotation classes
Note: we should move these classes to a common class as they will be reusable elsewhere

```python
# Bounding Box class  
class BoundingBox() :
    def __init__(self, xmin, ymin, xmax, ymax) :
        self.xmin = int(xmin)
        self.ymin = int(ymin)
        self.xmax = int(xmax)
        self.ymax = int(ymax)
        
    def to_string(self):
        return '[' + str(self.xmin) + ',' +  str(self.ymin) + ',' +  str(self.xmax) + ',' +  str(self.ymax) + ']' 
```

```python
# ObjectAnnotation class
class ObjectAnnotation() :
    '''
    Captures an annotation for an object in an image file. The bounding box and label 
    annotates 'where and what' about the object in the image
    '''
    def __init__(self, image_file, bounding_box, label) :
        self.image_file = image_file
        self.bounding_box = bounding_box
        self.label = label
    
    def to_string(self):
        return 'image_file = ' + self.image_file + ', bounding_box = ' + self.bounding_box.to_string() + ' label = ' + self.label
```

Convert the csv rows into list of annotation objects

```python
annotations=[]

# from the rows that we read from the csv file
# header : ['xmin', 'ymin', 'xmax', 'ymax', 'Frame', 'Label', 'Preview URL']
for row in rows :
    image_file = row[4] 
    bounding_box = BoundingBox(xmin=row[0], ymin =row[1], xmax = row[2], ymax=row[3])
    annotations.append(ObjectAnnotation(image_file = image_file, bounding_box=bounding_box, label = row[5]))
    
print(annotations[0].to_string())
```

```python
## group all annotations of an image by its file name into a dictionary

import collections

image_dict = collections.defaultdict(list)

for el in annotations:
    image_dict[el.image_file].append(el)
```

```python
for annotation in image_dict['1479498371963069978.jpg']:
    print(annotation.to_string())
```

```python
## Draw a sample image file withs its bounding boxes

sample_image = '1479498371963069978.jpg'

font = cv2.FONT_HERSHEY_SIMPLEX

img = mpimg.imread(IMAGE_DIR +sample_image)
annotes = image_dict[sample_image]
for el in annotes:
    cv2.rectangle(img,(el.bounding_box.xmin, el.bounding_box.ymin), (el.bounding_box.xmax, el.bounding_box.ymax), (0,0,255), 6)
    cv2.putText(img,el.label,(el.bounding_box.xmin,el.bounding_box.ymin-5), font, 1, (200,255,155), 13, cv2.LINE_AA)
plt.imshow(img)
```

<h2> Dataset Preparation</h2>

The single shot multibox detector algorithm that we implement here is detailed in Wei Liu's paper - https://arxiv.org/pdf/1512.02325.pdf. 

The image below is taken from Wei Liu's paper which details the SSD deep learning model
<p>
<img src="../doc_images/SSDModel.png" width="980" alt="SSD Model Image" />
</p>

To briefly explain the model, what we do here is use a pretrained VGG model and add additional convolution layers (starting from 'fc6' scaling down to 'conv_11_2') that capture more details about the image at different scales. We then take the outputs of specific conv layers (block4_conv3, fc7 etc.) aka feature maps in the architecture and for each of these layers we run a Convolution with a 3x3 filter of depth slice (num_default_boxes x (num_classes + 4)) and flatten it to produce an output y_layer_i of size (layer_i_width x layer_i_height x (num_default_boxes x (num_classes + 4 )). 

What this does is, for each cell in the layer, it creates default bounding boxes(the 4 is for the 4 points of the box) and gives probability scores for each class in each box. We repeat this for the other feature maps layers, flatten their outputs and combine them into this huge output list(y_pred) of bounding boxes with class probability scores for each cell in each output layer. By combining the outputs from different scaled layers we are actually learning the bounding at different scales. At training time, we will compare the predicted bounding boxes with ground truth bounding box and regress the predicted bounding box closer to the ground truth. 

This means that we need to build the y_ground_truth labels for training similar to the huge y_pred list. What we have is only the list of annotated bounding boxes for labeled objects. for e.g, if an image  has 5 vehicles, we will have 5 bounding boxes and their associated class id. What we need to do is for each cell in the image we need to come up with default bounding boxes(around 4 to 6 boxes with different aspect ratios) and associate the class id. If a bounding box for a cell falls on a car (say 90% of the bounding box is on the car) then we will associate car with the bounding box. If the bounding box falls on unknown class(background) then we tag at as background class. We do this for each cell of each pred layer to build the huge y_ground_truth for 1 input image

**Note:** Since the input image dimensions(300x300) are different from each layers dimensions we have to convert the absolute bounding box points to relative offsets so we can conistently identify where the vehichle should be in a particular layer's resolution.


**Trade-off**

Since y_ground_truth computation will be the same for different iterations during training, it makes sense to pre-compute the y_ground_truth ahead of time as part of dataset preparation to speed up training. But this means that we need to know the dimensions of the different prediction conv layers in the SSD model ahead of time. The downside is that the model can only work for one input image shape. If you decide to change the input shape then the model will have different dimensions at the pred conv layers. I think this is alright as we most likely will fix upon the input shape ahead of time (Wei uses 300x300 as input image).  The savings in training time outweighs the downside of static dimensions for the layers.


Note: We are using the VGG16 pretrained model from the Keras library. The dimensions of some of the layers are slightly off, for instance, block4_conv3 has a dimension 37x37 (whereas in the diagram above it is 38x38. We will have to generate the whole model first to fix upon the layer dimensions
